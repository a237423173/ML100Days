```python
import numpy as np

english_score = np.array([55,89,76,65,48,70])

math_score = np.array([60,85,60,68,np.nan,60])

chinese_score = np.array([65,90,82,72,66,77])

#上 3列共六位同學的英文、數學、國文成績
#請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略？
#第五位同學補考數學後成績為 55，請計算補考後數學成績平均、最大值、最小值、標準差？
#用補考後資料找出與國文成績相關係數最高的學科？

#各科成績平均
mean_eng = np.mean(english_score)
print(mean_eng)
mean_math = np.nanmean(math_score)
print(mean_math)
mean_ch = np.mean(chinese_score)
print(mean_ch)

#各科成績最大值
max_eng = np.max(english_score)
print(max_eng)
max_math = np.nanmax(math_score)
print(max_math)
max_ch = np.max(chinese_score)
print(max_ch)

#各科成績最小值
min_eng = np.min(english_score)
print(min_eng)
min_math = np.nanmin(math_score)
print(min_math)
min_ch = np.min(chinese_score)
print(min_ch)

#各科成績標準差
std_eng = np.std(english_score)
print(std_eng)
std_math = np.nanstd(math_score)
print(std_math)
std_ch = np.std(chinese_score)
print(std_ch)

math_score = np.array([60,85,60,68,55,60])

#補考後各科成績平均
mean_eng = np.mean(english_score)
print(mean_eng)
mean_math = np.mean(math_score)
print(mean_math)
mean_ch = np.mean(chinese_score)
print(mean_ch)

#補考後各科成績最大值
max_eng = np.max(english_score)
print(max_eng)
max_math = np.max(math_score)
print(max_math)
max_ch = np.max(chinese_score)
print(max_ch)

#補考後各科成績最小值
min_eng = np.min(english_score)
print(min_eng)
min_math = np.min(math_score)
print(min_math)
min_ch = np.min(chinese_score)
print(min_ch)

#補考後各科成績標準差
std_eng = np.std(english_score)
print(std_eng)
std_math = np.std(math_score)
print(std_math)
std_ch = np.std(chinese_score)
print(std_ch)

#相關係數
cor_ch_eng = np.corrcoef(chinese_score, english_score)
print(cor_ch_eng)
cor_ch_math = np.corrcoef(chinese_score, math_score)
print(cor_ch_math)
print("補考後國文成績與英文成績相關係數最高")
```
